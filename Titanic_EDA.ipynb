```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')
```

```python
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Sex'].value_counts()
df['Embarked'].value_counts()
df['Pclass'].value_counts()
```

```python
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.show()
```

Observation:

The Age distribution is right-skewed.

Most passengers were between 20-40 years old.

Very few passengers were above 60 years old

```python

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

```

Observation:

More people died (Survived = 0) compared to those who survived (Survived = 1).

Survival rate is lower than death rate among passengers.

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.show()
```


Observations:
- Median age of survivors is **lower** than non-survivors.
- Many older passengers (above 60) **did not survive**.
- Younger passengers (around 2030) have **higher survival rates**.
- Some outliers are visible among the survivors (extreme ages).


```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age Scatterplot')
plt.show()
```

Observations:
- Passengers who paid **higher fares** had a **higher chance of survival**.
- Most passengers who paid **low fares** (under $50) did not survive.
- A few younger passengers paid very high fares.
- There is **no strong linear relationship** between Age and Fare.


```python
sns.pairplot(df, hue='Survived')
plt.title('Pairplot - Survival Analysis')
plt.show()
```

Observations:
- Survivors tend to cluster in **higher fare and lower age** region.
- First-class passengers are highly represented among survivors.
- There is overlap between survivors and non-survivors, but **clear trends** are visible.
- Relationships among features are **non-linear and complex**.


```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=['int64', 'float64']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

 Observations:
- **Fare** and **Pclass** are **negatively correlated** (higher class = higher fare).
- **Survived** is **positively correlated** with **Fare** and **Pclass**.
- Age shows a **very weak correlation** with survival.
- SibSp and Parch (family members) have slight positive correlation.



 Final Summary of EDA on Titanic Dataset

- Majority of passengers were young adults between 20 and 40 years old.
- Survival chances were much higher for **females** compared to males.
- **First-class** passengers had better survival rates than second and third-class passengers.
- Passengers who paid higher **fares** were more likely to survive.
- There were missing values in the **Age**, **Cabin**, and **Embarked** columns.
- Heatmap showed that **Fare** and **Pclass** had strong negative correlation.
- Family members (SibSp, Parch) had slight positive influence on survival chances.
- Overall, **age, gender, ticket class, and fare** were important factors affecting survival.

This analysis highlights the importance of exploratory data analysis (EDA) in discovering patterns, anomalies, and insights before building machine learning models.
